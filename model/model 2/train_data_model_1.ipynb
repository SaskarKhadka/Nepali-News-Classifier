```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import codecs, json
import pickle
```

```python
with open('../../outputs/model 2/train_history.pkl', "rb") as hist:
    history = pickle.load(hist)
```

Now lets crete some plots from this data

```python
plt.plot(history['loss'])
plt.plot(history['val_loss'])
plt.title('Training and Validation Loss vs. Epochs')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.savefig("../../images/model 2/loss plot")
plt.show()
```

```python
plt.plot(history['accuracy'])
plt.plot(history['val_accuracy'])
plt.title('Training and Validation Accuracy vs. Epochs')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.savefig("../../images/model 2/accuracy plot")
plt.show()
```

```python
plt.plot(history['f1_score'])
plt.plot(history['val_f1_score'])
plt.title('Training and Validation F1 Score vs. Epochs')
plt.ylabel('f1 score')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.savefig("../../images/model 2/f1 score plot")
plt.show()
```

```python
avg_f1_scores_train = [sum(vals) / len(vals) for vals in history['f1_score']]
avg_f1_scores_val = [sum(vals) / len(vals) for vals in history['val_f1_score']]
```

```python
plt.plot(avg_f1_scores_train)
plt.plot(avg_f1_scores_val)
plt.title('Training and Validation F1 Score vs. Epochs')
plt.ylabel('f1 score')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.savefig("../../images/model 2/f1 score plot")
plt.show()
```

```python
plt.plot(history['precision'])
plt.plot(history['val_precision'])
plt.title('Training and Validation precision vs. Epochs')
plt.ylabel('precision')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.savefig("../../images/model 2/precision plot")
plt.show()
```

```python
plt.plot(history['recall'])
plt.plot(history['val_recall'])
plt.title('Training and Validation recall vs. Epochs')
plt.ylabel('recall')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.savefig("../../images/model 2/recall plot")
plt.show()
```
